# EDA- Haberman's Cancer Survival

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

df = pd.read_csv('haberman.csv')
df.head()
```

```python
df.columns = ["age", "operation_year", "axillary_lymph_node", "survival_status"]
```

```python
df.head()
```

Survival status (class Attribute)

* 1 = the patient survived 5 years or longer
* 2 = the patient died within 5 years

```python
df.shape
```

```python
# brief information about the data
print(df.info())
```

```python
df['survival_status'] = df['survival_status'].map({1:'Yes', 2:'No'})
df.tail()
```

```python
df.describe()
```

```python
df['survival_status'].value_counts()
```

```python
status_yes = df[df['survival_status'] == 'Yes']
status_yes.describe()
```

```python
status_no = df[df['survival_status'] == 'No']
status_no.describe()
```

# 1. Univariant Analysis - Histogram

```python
sns.FacetGrid(df,hue='survival_status',height=5)\
 .map(sns.distplot,'age')\
  .add_legend();
plt.show()
```

```python
sns.FacetGrid(df,hue='survival_status',height=5)\
 .map(sns.distplot,"operation_year")\
  .add_legend();
plt.show()
```

```python
sns.FacetGrid(df,hue='survival_status',height=5)\
 .map(sns.distplot,"axillary_lymph_node")\
  .add_legend();
plt.show()
```

# Multivariant analysis

```python
sns.jointplot(x='operation_year', y='age', data = df, kind='kde')
plt.show()
```
