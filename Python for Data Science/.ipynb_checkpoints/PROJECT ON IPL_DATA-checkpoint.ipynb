```python
# Loading important Libraries 
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Loading the ipl matchs dataset
ipl = pd.read_csv('Match.csv')
```

```python
# Getting the frequency of man of the match awards
ipl['ManOfMach'].value_counts()
```

```python
# Getting the frequency of TOP 10 man of the match awards
ipl['ManOfMach'].value_counts()[0:10]
```

```python
list(ipl['ManOfMach'].value_counts()[0:5].keys())
```

```python
plt.figure(figsize=(8,5))  # figsize=(x,y)
plt.bar(list(ipl['ManOfMach'].value_counts()[0:5].keys()),ipl['ManOfMach'].value_counts()[0:5], color='cyan' )
plt.show()
```

```python
# Getting the frequency of result column
ipl['Outcome_Type'].value_counts()
```

```python
ipl.head()
```

```python
# Extracting the records where team won batting first that mean won by runs only
batting_first = ipl[ipl['Win_Type'] != 'wickets']
batting_first
```

```python
# Find out the number of toss win w.r.t each matches
ipl['Toss_Winner'].value_counts()
```

```python
# we use histogram for numeric values
plt.figure(figsize=(8,5))
plt.hist(batting_first['Win_Margin'])
plt.xlabel("Runs")
plt.title('Distribution of runs')
plt.show()
```

```python
# Finding the number of wins w.r.t each team after batting first
batting_first['match_winner'].value_counts()
```

```python
# Making a bar-plot for top 3 teams with wins w.r.t each team after batting first
plt.figure(figsize=(10,5))
plt.bar(list(batting_first['match_winner'].value_counts().keys()[0:3]), list(batting_first['match_winner'].value_counts()[0:3]), color=['green', 'red', 'yellow'])
plt.show()
```

# WORKING WITH PIE-CHART

```python
# plt.pie(numeric, catagoric)
plt.figure(figsize = (6,5))
plt.pie(list(batting_first['match_winner'].value_counts()[0:10]), labels=list(batting_first['match_winner'].value_counts().keys()[0:10]), autopct = "%0.1f%%" , radius = 1.7)
plt.pie([1], colors = 'white', radius = 0.7)
plt.show()
```
