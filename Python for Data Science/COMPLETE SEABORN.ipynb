```python
import seaborn as sns
from matplotlib import pyplot as plt
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.lineplot( x = 'timepoint', y = 'signal',data = fmri)
plt.show()
```

```python
sns.lineplot(x = 'timepoint', y = 'signal', data = fmri, hue = 'event')
plt.show()
```

```python
sns.lineplot(x = 'timepoint', y = 'signal', data = fmri, hue = 'event', style = 'event')
plt.show()
```

```python
sns.lineplot(x = 'timepoint', y = 'signal', data = fmri, hue = 'event', style = 'event', markers = True)
plt.show()
```

# SEABORN BAR PLOT

```python
import pandas as pd
datas = pd.read_csv('Data.csv')
sns.barplot(x = 'Chemistry', y = 'physics', data = datas , hue = 'Name')
sns.set(style = 'whitegrid')
plt.show()
```

```python
sns.barplot(x= 'Chemistry', y = 'physics', data = datas ,palette = 'Blues_d')
plt.show()
```

```python
sns.barplot(x='Chemistry', y = 'physics', data = datas , palette = 'vlag')
plt.show()
```

```python
sns.barplot(x='Chemistry', y = 'physics', data = datas ,palette = 'rocket')
plt.show()
```

```python
sns.barplot(x='Chemistry', y = 'physics', data = datas ,color = "orange")
plt.show()
```

# scatter plot

```python
sns.scatterplot(x = 'Chemistry', y = 'physics', data = datas, hue = 'Name')
plt.show()
```

# SEABORN HISTOGRAM / DISTPLOT

```python
datas.head()
```

```python
sns.distplot(datas['Chemistry']) 
plt.show()
```

```python
sns.distplot(datas['Chemistry'], hist=False, color = 'purple') # for just frequency curve use hist=False
plt.show()
```

```python
sns.distplot(datas['Chemistry'],  kde = False , bins = 2)# use kde= False , for remove frequency curve
plt.show()
```

```python
sns.distplot(datas['Chemistry'],  kde = False , bins = 2, vertical = True)# use vertical = True , for shape into vertical
plt.show()
```

# Seaborn JointPlot

```python
sns.jointplot(x = "Chemistry", y = "physics", data = datas )
plt.show() # this gives us scatter and histogram as well 
```

```python
sns.jointplot(x = "Chemistry", y = "physics", data = datas , kind = 'reg') #here we add kind='reg' for curve line
plt.show() # this gives us scatter and histogram as well 
```

# SEABORN BOXPLOT

```python
sns.boxplot(x="Chemistry", y="physics", data = datas, linewidth = 3)
plt.show()
```

# SEABORN PAIRPLOT

```python
sns.pairplot(datas, hue = 'Chemistry')
plt.show()
```
